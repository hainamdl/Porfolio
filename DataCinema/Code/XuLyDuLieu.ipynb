```python
def XuLySoVe(df):
  if df["slot type"] == "ĐÔI":
    return 2*(df["total"]/df["ticket price"])
  else:
    return (df["total"]/df["ticket price"])
```

```python
df_TicKet['day_of_week'] = df_TicKet['date'].dt.day_name()
```

```python
def week_of_month(dt):
    first_day = dt.replace(day=1)
    dom = dt.day
    adjusted_dom = dom + first_day.weekday()
    return int(adjusted_dom / 7) + 1
```

```python
df_TicKet['date'].apply(week_of_month)
```
